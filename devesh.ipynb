### Question 2
I have used the Airlines dataset, which presents the flights from all USA airports for the period from 1990-2009. 
The attributes are origin, destination, number of passenger and date.
For simplicity, we will use the data of only 1 year.

```python
# Load the data
airlines = pd.read_csv('Airports2.csv')
```

```python
airlines.head()
```

```python
# Creating Nodes and Edges
airlines['year']= airlines.Fly_date.apply(lambda x: x.split('-')[0])
airlines_2008 = airlines[airlines.year=='2008']
airlines_routes=airlines_2008.groupby(['Origin_airport', 'Destination_airport']).sum()['Passengers'].reset_index()
#Created the airline routes using the passenger information
airlines_routes_valid = airlines_routes[airlines_routes.Passengers>1000]
all_nodes = airlines_routes_valid.Origin_airport.append(airlines_routes_valid.Destination_airport).unique()
all_edges = airlines_routes_valid[['Origin_airport', 'Destination_airport']]
```

```python
G = nx.Graph()
G.add_nodes_from(all_nodes)
G.add_edges_from(all_edges.as_matrix())
graph_pos = nx.spring_layout(G)
```

```python
org_df = airlines.groupby(['Origin_airport', 'Origin_city', 'Org_airport_lat','Org_airport_long', 
                                 'Origin_population']).count().reset_index()[['Origin_airport', 'Origin_city', 'Org_airport_lat','Org_airport_long','Origin_population']]
dest_df = airlines.groupby(['Destination_airport', 'Destination_city', 'Dest_airport_lat', 'Dest_airport_long', 'Destination_population']).count().reset_index()[['Destination_airport', 'Destination_city', 'Dest_airport_lat', 'Dest_airport_long', 'Destination_population']]
```

```python
airlines.groupby(['Origin_airport', 'Origin_city', 'Org_airport_lat','Org_airport_long', 
                                 'Origin_population'])


all_nodes_info= airlines.groupby(['Origin_airport', 'Origin_city', 'Org_airport_lat','Org_airport_long', 
                                 'Origin_population']).count().reset_index()[['Origin_airport', 'Origin_city', 'Org_airport_lat','Org_airport_long','Origin_population']]
```

```python
flight_sum_dict = {}
for i , j in sum_of_flights.iterrows():
    code= j['Origin_airport']
    city= j['Origin_city']
    lat = j['Org_airport_lat']
    long = j['Org_airport_long']
    flight = j['Flights']
    flight_sum_dict[code] =flight
```

```python
lat_dict = {}
long_dict={}
city_name_dict ={}
population_dict={}
for i , j in all_nodes_info.iterrows():
    code= j['Origin_airport']
    city= j['Origin_city']
    lat = j['Org_airport_lat']
    long = j['Org_airport_long']
    pop = j['Origin_population']
    
    lat_dict[code] =lat
    city_name_dict[code] =city
    long_dict[code] = long
    population_dict[code]=pop
```

```python
all_city_name=[]
for node in all_nodes:
    all_city_name.append(city_name_dict.get(node))
```

```python
all_pop_name=[]
for node in all_nodes:
    all_pop_name.append(population_dict.get(node))
```

```python
communities = community.best_partition(G)
```

```python
import plotly.offline as py
from plotly.graph_objs import *
py.init_notebook_mode(connected=True)
import numpy as np
import networkx as nx
%matplotlib inline
```

```python
edge_trace = Scatter(
    x=[],
    y=[],
    text=[],
    line=Line(width=0.05,color='#F00'),
    hoverinfo='text',
    mode='lines')

for edge in G.edges():
    x0, y0 = long_dict.get(edge[0], 39.94269943), lat_dict.get(edge[0],-87.611396789)
    x1, y1 = long_dict.get(edge[1], 39.94269943), lat_dict.get(edge[1],-87.611396789)
    x_trace = list(np.linspace(x0, x1, 10))
    y_trace = list(np.linspace(y0, y1, 10))
    x_trace.append(None)
    y_trace.append(None)
    edge_trace['x'] += x_trace
    edge_trace['y'] += y_trace


node_trace = Scatter(
    x = [],
    y = [],
    text=[],
    mode='markers',
    hoverinfo='text',
    marker=Marker(
        showscale=True,
        colorscale='Jet',
        reversescale=True,
        color=[],
        size=[],
        colorbar=dict(
            thickness=15,
            title='Community',
            xanchor='left',
            titleside='right'
        ),
        line=dict(width=2)))

for node in G.nodes():
    x, y = graph_pos[node]
    node_trace['x'].append(long_dict.get(node, 'SVN'))
    node_trace['y'].append(lat_dict.get(node, 'SVN'))
    node_trace['marker']['size'].append(np.log(flight_sum_dict.get(node, 197049)))
    node_trace['text'].append('Flights: '+str(flight_sum_dict.get(node, 197049))+' '+str(city_name_dict.get(node,'City')))
    node_trace['marker']['color'].append(communities[node])
    
```

```python
layout = dict(
        title = 'USA airport with Routes, Size represent number of flights, Colored by Communities<br>(Hover for airport names)',
        #colorbar = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    xaxis=dict(
        range=[-130, -65]
    ),
    yaxis=dict(
        range=[25, 50]
    )
    )
```

### Plotting communities in the Graph

```python
fig = Figure(data=Data([edge_trace, node_trace]),layout=layout)
#plt.plot(fig)
py.iplot(fig)
```

```python
py.offline.plot(fig, filename = 'filename.html', auto_open=False)
```

We can see that Cities in same or nearby states are grouped together. For example, for community =2, most of the cities are from Texas. viz : Midland, Lubbock, El Paso, Tulsa, Austin, Houston.
